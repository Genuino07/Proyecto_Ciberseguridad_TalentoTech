```python
import pandas as pd
import sqlite3


# Definir la ruta al archivo de la base de datos (que es un archivo .db con extensión .sql)
db_file_path  = "/content/drive/MyDrive/bootcamp/Proyecto/ciberseguridad.db"

# 1. Crear una conexión a la base de datos directamente desde el archivo
conn = sqlite3.connect(db_file_path)
cursor = conn.cursor()

# 2. Obtener la lista de tablas disponibles en esta base de datos
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Tablas disponibles en la base de datos:", [tbl[0] for tbl in tables])

# 3. Cargar datos en un DataFrame de pandas desde la base de datos
# Ahora que sabemos las tablas, podemos seleccionar una existente.
# Por ejemplo, si 'comunidades_energeticas' está en la lista de tablas:
if 'FROM amenazas' in [tbl[0] for tbl in tables]:
    query = "SELECT * FROM amenazas"
elif tables: # Si hay al menos una tabla, usa la primera
    query = f"SELECT * FROM {tables[0][0]}"
else:
    raise ValueError("No se encontraron tablas en el archivo .db.")

data = pd.read_sql_query(query, conn)

# Opcional: mostrar las primeras filas del DataFrame
print(data.head())

# 4. Cerrar la conexión
conn.close()

# 1. Crear una conexión a la base de datos directamente desde el archivo
conn = sqlite3.connect(db_file_path)
cursor = conn.cursor()

# 2. Obtener la lista de tablas disponibles en esta base de datos
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Tablas disponibles en la base de datos:", [tbl[0] for tbl in tables])

# 3. Cargar datos en un DataFrame de pandas desde la base de datos
# Ahora que sabemos las tablas, podemos seleccionar una existente.
# Por ejemplo, si 'comunidades_energeticas' está en la lista de tablas:
if 'FROM amenazas' in [tbl[0] for tbl in tables]:
    query = "SELECT * FROM  amenazas"
elif tables: # Si hay al menos una tabla, usa la primera
    query = f"SELECT * FROM {tables[0][0]}"
else:
    raise ValueError("No se encontraron tablas en el archivo .db.")

data = pd.read_sql_query(query, conn)

# Opcional: mostrar las primeras filas del DataFrame
print(data.head())

# 4. Cerrar la conexión
conn.close()
```
