```python
# Paso 2.1: Cargar los DataSets desde archivo en Formato .CSV

# Carga la tabla de hechos
Fact_SecurityIncidents = pd.read_csv('Fact_SecurityIncidents.csv')

# Carga la tabla de dimensiones
Dim_Location = pd.read_csv('Dim_Location.csv')
Dim_Industry = pd.read_csv('Dim_Industry.csv')
Dim_Attack = pd.read_csv('Dim_Attack.csv')
Dim_Defense = pd.read_csv('Dim_Defense.csv')
Dim_Time = pd.read_csv('Dim_Time.csv')

```

```python
# Paso 4:
# Contruir el DW (Integrando las tablas)

dw = (
    Fact_SecurityIncidents  # Tu tabla de hechos principal

    .merge(Dim_Location, on='Country_ID', how='left')

    .merge(Dim_Industry, on='Industry_ID', how='left')

    .merge(Dim_Attack, on='Attack_ID', how='left')

    # Unimos con la dimensión de Defensa
    .merge(Dim_Defense, on='Defense_ID', how='left')

    # Unimos con la dimensión de Tiempo (Año)
    .merge(Dim_Time, on='Time_ID', how='left')
)

# Mostrar el resultado final integrado
dw.head()

```
